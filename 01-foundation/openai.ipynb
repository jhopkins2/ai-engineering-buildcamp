```python
from openai import OpenAI
from dotenv import load_dotenv, dotenv_values

load_dotenv()
openai_client = OpenAI()
```

```python
response = openai_client.responses.create(model="gpt-4o-mini", input="Tell me a joke about Cyberpunk")
```

```python
print(response.output_text)
```

```python
messages = [
    {"role":"user", 
    "content":"Give me a bible verse from the King James Bible from the book of Jeremiah"
    }
]

response = openai_client.responses.create(
    model="gpt-4o-mini",
    input=messages
)
```

```python
print(response.output_text)
```

```python
response.usage
```

```python
MODEL_PRICES = {
    "gpt-4o-mini": {"input": 0.15, "output": 0.60},
    "gpt-5-nano": {"input": 0.075, "output": 0.30},
    "gpt-5-mini": {"input": 0.25, "output": 2.00},
    "gpt-5.2": {"input": 1.75, "output": 14.00},
    "gpt-5.2-pro": {"input": 21.00, "output": 168.00},
}

def calculate_cost(model_name: str, input_tokens: int, output_tokens: int) -> float:
    prices = MODEL_PRICES[model_name.lower()]
    input_cost = (input_tokens / 1_000_000) * prices["input"]
    output_cost = (output_tokens / 1_000_000) * prices["output"]
    return input_cost + output_cost
```

```python
usage = response.usage

cost = calculate_cost("gpt-4o-mini", usage.input_tokens, usage.output_tokens)

print(f"Cost: ${cost:.6f}")
```

```python
messages = [
    {
        "role":"system",
        "content":"you are an old, wise, travelling bard. You tell stories to auidences in taverns, campfires, etc. Use a spooky fantasy style to retell the stories." 
    },    
    {
        "role":"user",
        "content":"Tell me a sword and sorcery story about a classic anti-hero warrior trying to survive in a cruel world where magic is danger and trust is limited",
    }
]

stream = openai_client.responses.create(
    model="gpt-4o-mini",
    input=messages,
    stream=True
)

response = None
for event in stream:
    if hasattr(event, 'delta'):
        print(event.delta, end='')
    if hasattr(event, 'response'):
        response = event.response
```

```python
print(response.usage)
```

```python
usage = response.usage

cost = calculate_cost("gpt-4o-mini", usage.input_tokens, usage.output_tokens)

print(f"Cost: ${cost:.6f}")
```
